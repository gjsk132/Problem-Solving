> 문제 요약



### 1차 시도 : 브루트포스 (시간 초과)

```python
def solution(k, d):
    
    answer = 0
    
    for i in range(0, d+1, k):
        for j in range(0, d+1, k):
            if pow(d,2) < pow(i, 2)+pow(j,2):
                break
            else:
                answer += 1
    
    return answer
```

```python
k = 2
d = 4
```

```python
solution(k, d)
```

### 2차 시도 : 수학 이용 > 피타고라스 정리

```python
import math

def solution(k, d):
    
    answer = 0
    
    for i in range(0, d+1, k):
        length = int(math.sqrt(pow(d, 2)-pow(i, 2)))
        answer += (length//k) + 1
    
    return answer
```
