```python
from collections import defaultdict

def solution(s):
    cnt = defaultdict(int)
    for value in list(map(int, s.replace("{", "").replace("}", "").split(","))):
        cnt[value] += 1

    answer = sorted(cnt.items(), key=lambda x:-x[1])

    answer = [k for k, v in answer]

    return answer
```

```python
solution("{{4,2,3},{3},{2,3,4,1},{2,3}}")
```
