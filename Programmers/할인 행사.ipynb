> 문제 요약
- 일정 금액 지불 시 10일 동안 회원 자격 부여
- 매일 한 가지 제품을 할인
- 할인 제품은 하루에 하나씩만 구매 가능

- 원하는 제품과 수량이 할인하는 날짜와 10일 연속으로 일치할 경우에 맞춰 회원가입

- 그리디 (단순 빡 구현인듯)

```python
def solution(want, number, discount):
    left_shopping_list = {x : y for x, y in zip(want, number)}

    can_sign_up = 0

    print(len(discount))

    for day, dc in enumerate(discount):
        print(day)
        if dc in want:        
            left_shopping_list[dc] -= 1

        if day < 9:
            continue
        elif not day == 9:
            finish_dc = discount[day-10]

            if finish_dc in want:
                left_shopping_list[finish_dc] += 1
        
        left_cnt = sum([max(cnt, 0) for cnt in left_shopping_list.values()])

        if left_cnt == 0:
            can_sign_up += 1

    return can_sign_up
    
```

```python
want = ["banana", "apple", "rice", "pork", "pot"]
number = [3, 2, 2, 2, 1]
discount = ["apple", "banana", "rice", "apple", "pork", "banana", "pork", "rice", "pot", "banana", "apple", "banana", "chicken", "apple"]
solution(want, number, discount)
```
