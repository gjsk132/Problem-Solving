```python
strs = ["flower","flow","flight"]

def longestCommonPrefix(self, strs: List[str]) -> str:
    strs.sort(key=lambda x: len(x))
    answer = ""

    std = strs[0]

    for std_pos, std_alpha in enumerate(std):
        for word in strs[1:]:
            if word[std_pos] == std_alpha:
                continue
            
            return answer
                    
        answer = std[:std_pos+1]

    return answer

                     
longestCommonPrefix(strs)
```
