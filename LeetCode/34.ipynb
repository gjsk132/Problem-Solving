LeetCode 34 : Find First and Last Position of Element in Sorted Array

```python
def searchRange(nums, target):
    answer = [-1, -1]
    if target in nums:
        pos = nums.index(target)
        answer[0] = pos
        pos += 1

        while nums[pos] == target:
            if pos == len(nums):
                break
            pos += 1
        
        answer[1] = pos-1

    return answer
```

```python
searchRange([5,7,7,8,8,10], 5)
```
