> 문제 요약

```python
from collections import defaultdict
from heapq import *

input = open("input.txt").readline

tree = defaultdict(list)

for i in range(int(input())):
    root, lnode, rnode = input().split()
    tree[root] = [lnode, rnode]

def preorder_traverse(root):
    ln, rn = tree[root]

    print(root, end="")

    if not ln == '.':
        preorder_traverse(ln)

    if not rn == '.':
        preorder_traverse(rn)

def inorder_tranverse(root):
    ln, rn = tree[root]

    if not ln == '.':
        inorder_tranverse(ln)

    print(root, end="")

    if not rn == '.':
        inorder_tranverse(rn)

def postorder_tranverse(root):
    ln, rn = tree[root]

    if not ln == '.':
        postorder_tranverse(ln)

    if not rn == '.':
        postorder_tranverse(rn)

    print(root, end="")

preorder_traverse('A')
print()

inorder_tranverse('A')
print()

postorder_tranverse('A')
print()
```
