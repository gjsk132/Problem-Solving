백준 26093 : 고양이 목에 리본 달기

[[바로가기]](https://www.acmicpc.net/problem/26093)

> 문제 요약
- N마리의 고양이에게 K종류의 리본 달기
- 리본의 개수는 무한
- 각 고양이의 리본 선호도 > 만족도

- 번호순으로 한줄 >> 리본달기
- 이웃한 고양이와 같은 종류의 리본 다는 것 X
- 최대한의 만족도 합

> 주어진 조건
- N(고양이 수), K(리본 수)

> 문제 유형
- DP

> 문제 풀이
- 0 ... K-1 index : 각 색상을 착용할 때, 될 수 있는 가장 큰 값
- 바로 이전의 데이터만 사용해도 괜찮음! (DP[2][K]의 크기를 돌릴 예정)
- 시간초과 ( 메모리는 괜찮을듯 )


- 왜 시간 초과일까?
- 문제 다시 읽기
- 연속으로 같은 리본을 달지 않으면 된다.

- 가장 큰 값과 2번째로 큰 값, 리본 색깔의 index만 기억하기?
- 그리디?


```python
input = open('input.txt').readline

cat, ribbon = map(int,input().split())

dp = [[0]*ribbon for _ in range(2)]

for cnt in range(cat):
    
    dp[cnt%2] = list(map(int,input().split()))

    for color in range(ribbon):
        dp[cnt%2][color] += max(dp[cnt%2-1][:color]+dp[cnt%2-1][min(color+1,ribbon):])
    
    print(dp[cnt%2])

print(max(dp[cat%2-1]))
```

```python
input = open('input.txt').readline

cat, ribbon = map(int,input().split())

dp = [0*ribbon] + [list(map(int,input().split()))]

for cnt in range(cat-1):

    max_first, mf_idx = 0, -1
    max_second = -1

    dp[cnt%2] = list(map(int,input().split()))

    for idx, perf in enumerate(dp[cnt%2-1]):
        if max_first < perf:
            max_second = max_first
            mf_idx = idx
            max_first = perf
        elif max_second < perf:
            max_second = perf
    
    for idx in range(ribbon):
        dp[cnt%2][idx] += max_first if idx != mf_idx else max_second

print(max(dp[cat%2]))
```
