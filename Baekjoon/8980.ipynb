백준 8980 : 택배

[[바로가기]](https://www.acmicpc.net/problem/8980)

> 문제 요약
- 버스는 1에서 N가지 한 번 이동한다.
- 이동하면서 보내는 마을과 받는 마을이 정해진 박스를 박스 용량에 맞춰서 실어보낸다.
- 최대한 많은 박스를 옮기는 방법을 찾아라.


> 입력
- 마을 수, 트럭 용량
- 2 <= N(마을 수) <= 2,000
- 1 <= C(트럭 용량) <= 10,000
- 1 <= M(박스 정보 수) <= 10,000
- M 만큼, 보내는 마을번호, 받는 마을번호, 보내는 박스 수
- 박스 수는 1 ~ 10,000이다.

흐음... 회의실 배정, 강의실이랑 많이 비슷하게 느껴짐.

방법 1 

빨리 택배를 담고, 빨리 택배를 내리는게 중요할 것 같음 | 힙 사용 (택배를 내릴 때마다 )

[ 추측 ]
정렬 + 힙 문제

방법 2

택배를 담고, 내리는 간격 작은 순으로 먼저 넣기

간격이 1이라면? 사실상 택배의 용량만큼 계속해서 옮길 수 있다.


```python
#  방법 1 : 정렬 + 힙
#  15% 정답
input = open("input.txt").readline

node, max_box = map(int,input().split())
box_info = [tuple(map(int,input().split())) for _ in range(int(input()))]

box_info = sorted(box_info, key=lambda x:(x[1],x[0]))

now_box = 0

total_box = 0

from heapq import *

move_box = []

for start, end, boxes in box_info:
    while move_box and move_box[0][0] <= start:
        now_box -= heappop(move_box)[1]

    can_move = min(max_box-now_box, boxes)
    heappush(move_box, (end, can_move))
    now_box += can_move
    total_box += can_move

print(total_box)
```
