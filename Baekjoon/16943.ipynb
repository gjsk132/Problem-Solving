백준 16943 : 숫자 재배치

[[바로가기]](https://www.acmicpc.net/problem/16943)

> 문제 요약

- 숫자 A에 있는 각 자리 숫자를 이용하여 숫자 B보다 작지만 가장 큰 수 출력 (없으면 -1)

> 방법

그냥 단순히 모든 경우의 수를 구하고, max를 출력하는 방식.

숫자는 최대 9자리 숫자 -> 그럼 9개를 골라야하는데... 모든 경우의 수를 구하면



```python
input = open("input.txt").readline

from collections import defaultdict

given_num, limit = map(int,input().split())
num_cnt = defaultdict(int)
num_len = 0

for num in str(given_num):
    num_cnt[num] += 1
    num_len += 1

num_kinds = sorted(num_cnt.keys(), reverse=True)
num_combs = []

def backtracking():
    if len(num_combs) == num_len:
        return int("".join(map(str, num_combs)))
    
    for n in num_kinds:
        if num_cnt[n] == 0:
            continue

        num_combs.append(n)
        num_cnt[n] -= 1
        result = backtracking()
        
        if result and result < limit:
            return result
        
        num_combs.pop()
        num_cnt[n] += 1

answer = backtracking()
print(-1 if len(str(answer)) != num_len else answer)
```

```python
# 포기한 코드짜기... ( 뭔가 구현으로 풀 수 있을 것 같았는데, 뭔가 까다로움. )

input = open("input.txt").readline

from heapq import *

given_num, limit = input().split()
nums = sorted(list(given_num), reverse=True)

result = ""

step = 0
not_find = True

while nums and not_find:
    for i in range(len(nums)):
        if nums[i] <= (lmt := limit[step]):
            n = nums[i]
            break

    if n < lmt:
        result += n
        nums.remove(n)
        result += "".join(nums)

    elif n == lmt:
        nums.remove(n)
        if "".join(nums) >= "".join(limit[step+1:]):
            if i == len(nums):
                result = "-1"
                break
            result += nums[i] + n
            
            nums.remove(nums[i])
            result += "".join(nums)
        else:
            result += n
            step += 1
            continue

    else:
        result = "-1"
        
    break

if result[0] == '0' or int(result) >= int(limit):
    print(-1)
else:
    print(result)
```
