> 문제 요약
- a의 위치와 b의 위치가 주어진다.
- a에서 c까지의 거리와 b에서 c까지의 거리가 주어진다.
- c가 될 수 있는 위치의 경우의 수를 구하여라.

> how?
- 각 지점에서 반지름이 n인 원을 그렸을 때의 교점을 생각하면된다.

1. 교점이 0인 경우는 2가지이다.
- 두 원이 분리되어 있는 경우
- 하나의 원 안에 원이 겹치지 않고, 포함되어 있는 경우

2. 교점이 1인 경우는 1가지이다.
- 두 원이 한 지점에서 접하는 경우

3. 교점이 무수히 많은 경우는 1가지이다.
- 두 원이 완전히 겹치는 경우

4. 교점이 2인 경우는 위에 3 종류를 제외하고 남은 원들이다.

```python
input = open(0).readline

def count_intersections():
    x1, y1, r1, x2, y2, r2 = map(int, input().split())

    ctr_dist = pow(pow(x1-x2, 2) + pow(y1-y2, 2), 1/2)
    
    if x1 == x2 and y1 == y2 and r1 == r2:
        return (-1 if r1 !=0 else 1)
    elif ctr_dist > r1 + r2 or ctr_dist + min(r1, r2) < max(r1, r2):
        return 0
    elif ctr_dist == r1 + r2 or ctr_dist + min(r1, r2) == max(r1, r2):
        return 1
    else:
        return 2
    
for _ in range(int(input())):
    print(count_intersections())
```
