백준 1085 : 직사각형에서 탈출

[[바로가기]](https://www.acmicpc.net/problem/1085)

> 문제 요약
- 현 위치 ( x, y )
- 각 변이 좌표축에 평행
- 왼쪽 아래 꼭짓점 ( 0, 0 )
- 오른쪽 위 꼭짓점 ( w, h )
- 직사각형의 경계선까지 가는 거리의 최솟값

한 지점에서 (0~w, 0~h)중 존재하는 (x, y)

여기서 직사각형 경계까지의 최솟값 구하기

> 해결
- 각 변에 떨어진 거리를 구하고 그 중에서 최소값을 구한다.

```python
input = open("input.txt").readline

x, y, w, h = map(int, input().split())
print(min(x, w-x, y, h-y))
```
