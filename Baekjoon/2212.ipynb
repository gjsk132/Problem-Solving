백준 2212 : 센서

[[바로가기]](https://www.acmicpc.net/problem/2212)

> 문제 요약
> - 집중국 위치가 주어지고, 집중국 위치에 센서를 설치한다.
> - 수신 가능 영역의 갯수가 주어질 때
> - 수신 가능 영역의 넘위를 **최소값**을 출력해야 한다.

```python
input = open("input.txt").readline

node = int(input())
senser = int(input())

node_info = sorted(map(int,input().split()))
gap_info = sorted([node_info[i+1]-node_info[i] for i in range(node-1)])

print(sum(gap_info[:node-senser]))
```
