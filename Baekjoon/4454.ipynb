백준 4454 : 상근이의 여자친구

[[바로가기]](https://www.acmicpc.net/problem/4454)

> 문제 요약

- 기름의 양과 거리가 주어진다.
- 차의 속도에 따라 소모되는 기름의 양을 나타내는 av^4 + bv^3 + cv^2 + dv가 있다.
- 속도가 일정할 때, 기름이 바닥나지 않으면서 빨리 도착하려면 시속 몇으로 가야하는가?


> 생각 / 주의할 점

- 테스트 케이스의 수가 나와있지 않는다. (입력값이 없을 때까지 실행)

- float값을 변수로 받음.


> av^4 + bv^3 + cv^2 + dv


```python
input = open("input.txt").readline

def binary_search(start, end, cnt=0):

    if int(start*100) == int(end*100):
        return str(start)

    mid = (start + end)/2

    total_oil = ((a*mid**3 + b*mid**2 + c*mid + d) * m)

    if total_oil <= t:
        start = mid
    else:
        end = mid

    return binary_search(start, end, cnt+1)

while (string := input()):
    try:
        a, b, c, d, m, t = map(float, string.split())
        num, decimal = map(str, binary_search(0, 1000).split('.'))
        print("{}.{}".format(num, decimal[:2]))
    except EOFError:
        break
```
