백준 1374 : 강의실

[[바로가기]](https://www.acmicpc.net/problem/1374)

> 문제 요약
> - 모든 강의의 시작, 끝 시간이 있음.
> - 강의 시간들이 주어질 때, 최대한 적은 강의실을 사용하는 경우를 구하는 문제

> 입력
> - 강의 수
> - 강의의 수만큼 강의의 숫자, 강의의 시작과 끝 시간
> - 시간이 0부터 10억 (무조건 터진다.)
> - 강의 수 100,000 => 강의 수를 기준으로 해야한다는 것을 알 수 있음.

이 문제를 단순히 초로 세려고 하면 안된다. ( 초가 10억까지 가기 때문에 시간 초과)

```python
from heapq import *

input = open("input.txt").readline

course_cnt = int(input())
course_time = []

for _ in range(course_cnt):
    idx, start, end = map(int,input().split())
    course_time.append((start, end))

course_time = sorted(course_time, key=lambda x:(x[0],x[1]))

end_time = []
max_cnt = 0

for start, end in course_time:
    while end_time and end_time[0] <= start:
        heappop(end_time)
    
    heappush(end_time, end)

    max_cnt = max(max_cnt, len(end_time))

print(max_cnt)
```
