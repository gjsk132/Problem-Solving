백준 2193 : 이친수

[[바로가기]](https://www.acmicpc.net/problem/2193)

> 문제 요약

1로 시작하고, 1이 연속 2회 이상 나오지 않는 이진수를 이친수라고한다.

```python
input = open("input.txt").readline

n = int(input())

answer = 0

if n <= 2:
    print(1)
else:
    memo = [1]*n
    
    for i in range(2, n):
        memo[i] += sum(memo[:i-1])

    print(memo[-1])
```
