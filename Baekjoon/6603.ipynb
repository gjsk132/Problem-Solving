> 문제 요약
- n개의 정수가 있는 집합이 주어질 때, 집합에서 6개의 숫자를 고를 수 있는 모든 경우의 수를 출력하는 문제이다.

- 설명 > 백트레킹!

```python
input = open("input.txt").readline

def backtracking(cnt, last):
    if cnt == 6:
        print(*comb)
        return
    
    for num in lotto:
        if num < last or num in comb:
            continue
        comb.append(num)
        backtracking(cnt+1, num)
        comb.pop()

while True:
    lotto = list(map(int,input().split()))

    if lotto.pop(0) == 0:
        break

    lotto.sort()
    comb = []

    backtracking(0, 0)
    print()
```
