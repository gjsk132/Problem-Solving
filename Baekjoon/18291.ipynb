백준 18291 : 비요뜨의 징검다리 건너기

[[바로가기]](https://www.acmicpc.net/problem/18291)

> 문제 요약

- 1 ~ N까지의 번호
- 그냥 1에서 N까지 갈 수 있는 모든 경우의 수를 구하는 문제

- 재귀 (1과 N을 제외한 나머지 숫자들 중 겹치지 않게 선택하기 + 1)

```python
# 그리디 -> 응 안돼 ( 메모리 초과 )

from itertools import permutations

input = open("input.txt").readline

for _ in range(int(input())):
    n = int(input())
    total = 0

    for step in range(1,n-2):
        total += len(list(permutations(range(n-2), step)))
        
    print(total+2)
```
