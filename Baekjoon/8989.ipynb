백준 8989 : 시계

[[바로가기]](https://www.acmicpc.net/problem/8989)

> 문제 요약
- 시침과 분침이 이루는 각도 중 가장 작은 각도 측정 ( 0 ~ 180 )
- 5개의 시간이 주어짐
- 중간값을 갖는 각도
- (근데 이게 세심한 각도까지 생각해야하는건가...?)

```python
input = open("input.txt").readline

for _ in range(int(input())):
    times = sorted(list(input().split()))

    angle = []

    for idx, t in enumerate(times):
        hour, minute = map(int, t.split(":"))
        hour = (hour*60 + minute)%720
        minute *= 12
        
        angle.append([min(abs(hour-minute), 720-abs(hour-minute)), idx])
        
    angle.sort()
    print(times[angle[2][1]])
```
