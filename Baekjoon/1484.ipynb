백준 1484 : 다이어트

[[바로가기]](https://www.acmicpc.net/problem/1484)

> 문제 요약

- a^2 - b^2 = G
- G가 주어지고 가능한 a(a > b)의 수를 구하는 문제이다.

> 문제 파악

- 하나의 a값에서 a^2와 b^2의 차이가 가장 작을 때는 b = a-1일 때이다.
- 해당 경우에서 두 제곱의 차는 a^2 - (a-1)^2 = (a-a+1)(a+a-1) = 2a-1이다.
- 2a-1가 G보다 커지게 되는 순간부터는 원하는 값이 절대 나올 수 없다.
- 그 결과 a의 최대값은 a = (G + 1)/2 라는 것을 알 수 있다.
- worst case >> a = 100,000+1/2 ->> a == 50,001 이다.

> 알고리즘

- 단순 그리디? 두 제곱의 차 > O(N^2) > 2,500,000,000 > 약 10초 (제한 : 8초)... 가능?
- 이거 찾을 때는 이분탐색


```python
input = open("input.txt").readline

gap = int(input())
max_weight = (gap)//2 + 1

def binary_search(start, end, num):
        
    if start > end:
        return 0
    
    mid = (start + end) // 2

    tmp = num**2 - mid**2

    if tmp == gap:
        return num
    elif tmp < gap:
        end = mid - 1
    else:
        start = mid + 1

    return binary_search(start, end, num)
    
ans = []

for weight in range(1, max_weight+1):
    if (result := binary_search(1, weight, weight)):
        ans.append(result)

if ans:
    print(*ans)
else:
    print(-1)
```

```python
input = open("input.txt").readline

gap = int(input())
max_weight = (gap)//2 + 1

low, high = 1, 2

ans = []

while high <= max_weight:
    tmp = pow(high, 2) - pow(low, 2)
    if tmp == gap:
        ans.append(high)
    if tmp > gap:
        low += 1
        continue       
    
    high += 1
        
if ans:
    print(*ans)
else:
    print(-1)
```
