백준 16401 : 과자 나눠주기

[[바로가기]](https://www.acmicpc.net/problem/16401)

> 문제 요약
- M명의 조카, N개의 과자 
- 1명에게 줄 수 있는 막대 과자의 최대 길이
- 부술 순 있지만, 합칠 순 없음

> 생각
- 일단 이분 탐색처럼 생긴 문제
- 10**9 부터 나누기 시작
- N명에게 나눠줄 수 있는 최대한의 과자 길이를 구해야 한다.

> 로직
- 이분 탐색
- 특정 길이 length에서 몇 개의 막대과자가 가능한지 구하기
- N명에게 줄 수 있는데... 사람 수가 많다면? > 길이를 늘리기
- 사람 수가 적다면? > 길이를 줄이기

- 위 과정을 반복한다.

```python
input = open("input.txt").readline

people, snacks = map(int, input().split())

snack = sorted(list(map(int, input().split())))
snack = snack[max(0,snacks-people):]

start = 1
end = snack[-1]

while start <= end:
    mid = (start + end)//2
    give_people = 0

    for s in snack:
        give_people += s//mid
    if give_people >= people:
        start = mid + 1
    else:
        end = mid - 1
    
print(end)
```
