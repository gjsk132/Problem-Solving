백준 14241 : 슬라임 합치키

[[바로가기]](https://www.acmicpc.net/problem/14241)

> 문제 요약
- 슬라임을 합치다가 1마리가 된다면 끝난다.
- 슬라임의 크기는 x+y
- 슬라임을 합칠 때의 점수 x*y

- heaqp를 통해서 큰 값을 계속해서 반복

```python
from heapq import *
input = open("input.txt").readline

cnt = int(input())
slime = [-plus for plus in map(int,input().split())]
heapify(slime)

score = 0

while (cnt := cnt-1):
    s1, s2 = heappop(slime), heappop(slime)
    heappush(slime, s1+s2)
    score += s1*s2

print(score)
```
