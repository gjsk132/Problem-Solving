백준 2240 : 자두나무

[[바로가기]](https://www.acmicpc.net/problem/2240)

> 문제 요약
- 1 또는 2가 번갈아가면서 주어진다.
- 현재 위치에 따라 1 > 2 // 2 > 1 로 바꿀 수 있는 횟수가 정해져있다.

- 시간은 1,000초 동안 떨어지고, 최대 30번만 움직일 수 있다고 한다.
- 최대 W이므로 이하 여도 괜찮을 듯

- 자두를 최대로 얻을 수 있는 갯수를 출력해라.

> 유형
- DP
- 고려할 사항을 생각해.

- 언제 나무 위치를 바꿀 것인가? ( 이것을 기준으로 한다. )
- 몇 번 위치를 바꾸었는가?
- 최대 몇 개까지 쌓여있는가?

- 어떤 식으로 값을 기억해야 할까?
- dp[현재까지의 time][변경 횟수]

- 이게 자리를 바꾸면 위치가 어딘지는 정해져있음
- idx가 짝수면 1이고, idx가 홀수면 2이다.

```python
input = open("input.txt").readline

time, can_change = map(int,input().split())

dp = [[0]*(can_change+1) for _ in range(time)]

for t in range(time):
    for set in range(int(input())-1, can_change+1, 2):
        dp[t][set] += 1

for t in range(1, time):
    for idx in range(min(t, can_change)+1):
        if idx == 0:
           dp[t][idx] += dp[t-1][idx]
           continue

        dp[t][idx] += max(dp[t-1][idx-1], dp[t-1][idx])

for _ in dp:
    print(_)

print(max(dp[-1]))
```
