백준 3009 : 네 번째 점

[[바로가기]](https://www.acmicpc.net/problem/3009)

> 문제 요약
- 주어진 세 점
- 축에 평행한 직사각형을 찾기 위한 네 번째 점 찾기

```python
input = open("input.txt").readline

x = []
y = []

for i in range(3):
    nx, ny = map(int, input().split())
    if nx in x:
        x.remove(nx)
    else:
        x.append(nx)
    if ny in y:
        y.remove(ny)
    else:
        y.append(ny)

print(x[0], y[0])
```
