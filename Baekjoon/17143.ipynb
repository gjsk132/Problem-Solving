백준 17143 : 낚시왕

[[바로가기]](https://www.acmicpc.net/problem/17143)

> 문제 요약
- 열을 한 칸씩 옮기면서 가장 위에 있는 상어를 잡는다.
- 잡은 상어 길이의 총합을 구하는 문제

> 생각
- 상어가 이동하는 걸 잘 짜야한다.
- 방향으로 이동하다가 범위 밖으로 넘어가면 반대로 가게한다.
- 겹치면 항상 큰 값이 남도록 한다.

---

- 이거 상어를 한 칸씩 이동시키면 안된다.
- 한 상어의 다음 위치를 지정할 수 있는 방법
- 움직인 상어가 아닌 다른 상어를 움직여야 한다. ( 다행이도 메모리 제한이 나름 넉넉함 )
- 2개의 테이블을 주고 받는 걸로 생각해보기

---
아니면 상어가 언제 잡아먹히는지를 체크하는건 어떨까?

총 100초 중에서 어느 위치에 있을지를 판단하는 거임

map에 잡아먹히는 위치와 값을 적는다.

---

일단 세로로 움직이는 애들을 먼저 파악하기 ( 무조건 한번만 낚시왕과 겹치게 된다. )

다음은 가로로 움직이는 애들 ( 땅에 가까운 순서로 ) ( 여러 번 낚시왕과 겹칠 수 있다. )

근데 중간에 잡아먹히는거 까지 고려하면 안될 것 같음
그냥 dq로 뺑뺑 돌리기 간다.

---

```python
from collections import deque

input = open("input.txt").readline

R, C, M = map(int,input().split())

shark_cnt = M
dq = deque([])

table = [[0]*C for _ in range(R)]

for _ in range(M):
    r, c, s, d, z = map(int, input().split())

    if not s:
        s = s % (2*R-2) if d <= 2 else s % (2*C-2)

    table[r-1][c-1] = max(table[r-1][c-1], z)
    dq.append((r-1, c-1, s, d, z))

result = 0

for i in range(C):
    
    # fishing
    for j in range(R):
        if not table[j][i]:
            continue
        
        result += table[j][i]
        table[j][i] = 0
        break

    # shark_check
    for _ in range(shark_cnt):
        r, c, s, d, z = dq.popleft()

        if not table[r][c] == z:
            shark_cnt -= 1
            continue
        else:
            table[r][c] = 0
            dq.append(( r, c, s, d, z ))

    # shark_move
    for _ in range(shark_cnt):
        
        r, c, s, d, z = dq.popleft()

        # 1 : 위 / 2 : 아래 / 3 : 오른쪽 / 4 : 왼쪽
        if d <= 2:
            r += -s if d==1 else s

            while not 0 <= r < R:
                r = abs(r) if r < 0 else 2*R-2-r
                d = 1 if d==2 else 2

        else:
            c += s if d == 3 else -s

            while not 0 <= c < C:
                c = abs(c) if c <0 else 2*C-2-c
                d = 3 if d == 4 else 4

        table[r][c] = max(table[r][c], z)
        dq.append((r, c, s, d, z))
    
print(result)
```
