백준 9095 : 1, 2, 3 더하기

[[바로가기]](https://www.acmicpc.net/problem/9095)

> 문제 요약
- 1, 2, 3을 더해서 다음 값을 만들 수 있는 경우의 수를 구하는 방법
( 순서가 달라지는 것도 카운트 한다. )
- 1 ) 단순히 경우의 수를 통해 구하는 방법
- 2 ) dp 사용하기 ( n이 11보다 작으니 충분히 가능 ) >> 순서가 겹치는 구간이 발생해서 안된다.
- 3 ) 미리 해두고, print하기... ㅎㅎ

```python
input = open("input.txt").readline

dp = [1 for _ in range(11)]

dp[0:4] = [0,1,2,4]

for i in range(4,11):
    dp[i] = dp[i-1] + dp[i-2] + dp[i-3]

for i in range(int(input())):
    num = int(input())
    print(dp[num])
```
